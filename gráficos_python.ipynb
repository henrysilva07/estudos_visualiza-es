# Criando um gr√°fico com matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Criando subplot

```python
anscombe = sns.load_dataset("anscombe")
```

```python
anscombe.dataset.unique()
```

```python
#Filtro o dataset
df_1 = anscombe[anscombe['dataset'] == 'I']
df_2 = anscombe[anscombe['dataset'] == 'II']
df_3 = anscombe[anscombe['dataset'] == 'III']
df_4 = anscombe[anscombe['dataset'] == 'IV']
```

```python
# Criando uma fig para plotar o gr√°fico
plt.plot(df_1['x'] , df_1['y']) # Por default √© utilizado gr√°fico de linha 
plt.show()
```

```python
# Criando uma fig para plotar o gr√°fico
plt.plot(df_1['x'] , df_1['y'], 'o') 
```

```python
# Criando uma figura 
fig = plt.figure( figsize =(10,10))
# Criando o subplot
axes_1 = fig.add_subplot(2,2,1)
axes_2 = fig.add_subplot(2,2,2)
axes_3 = fig.add_subplot(2,2,3)
axes_4 = fig.add_subplot(2,2,4)

#Criando o gr√°fico de cada subplot#
axes_1.plot(df_1['x'], df_1['y'], 'o')
axes_1.set_title("dataset_1")
axes_1.set_xlabel("Eixo X")
axes_1.set_ylabel("Eixo Y")
axes_2.plot(df_2['x'], df_2['y'], 'o')
axes_2.set_title("dataset_2")
axes_2.set_xlabel("Eixo X")
axes_2.set_ylabel("Eixo Y")
axes_3.plot(df_3['x'], df_3['y'], 'o')
axes_3.set_title("dataset_3") 
axes_3.set_xlabel("Eixo X")
axes_3.set_ylabel("Eixo Y")
axes_4.plot(df_4['x'], df_4['y'], 'o')
axes_4.set_title("dataset_4")
axes_4.set_xlabel("Eixo X")
axes_4.set_ylabel("Eixo Y")

fig.suptitle("Ascombe Data")
fig.tight_layout()
```
